```python
import json
import requests
import os

from dotenv import load_dotenv

load_dotenv()

# Set the base URL for the Pocket Casts API
url = "https://api.pocketcasts.com/user/login"

# Set your Pocket Casts username and password
email = os.environ['POCKETCASTS_EMAIL']
password = os.environ['POCKETCASTS_PASSWORD']

# Set the HTTP headers for the request
headers = {
    "content-type": "application/json",
}

# Set the request parameters
payload = json.dumps({
  "email": email,
  "password": password,
  "scope": "webplayer"
})

# Send the request to the Pocket Casts API
response = requests.post(url, headers=headers, data=payload)

# Set new HTTP headers using Bearer token
headers = {
    'Authorization': f"Bearer {response.json()['token']}"
}

url = "https://api.pocketcasts.com/user/history"

response = requests.post(url, headers=headers)
response.json()
```
